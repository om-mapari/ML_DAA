```python
# f) Now extract only Age, Sex, ChestPain, RestBP, Chol. Randomly divide dataset in training
# (75%) and testing (25%).
dataset = data[["Age", "Sex", "ChestPain", "RestBP", "Chol"]] 
training_data, testing_data = train_test_split(dataset,test_size=0.25);

print("training_data => ")
print(training_data)
print("testing_data => ")
print(testing_data)
```

```python
# Through the diagnosis test I predicted 100 report as COVID positive, 
# but only 45 of those were actually positive. 
# Total 50 people in my sample were actually COVID positive. 
# I have total 500 samples.
# Create confusion matrix based on above data and find

# I. Accuracy
# II. Precision
# III. Recall
# IV. F-1 score
```

```python
actualData = [];

predictedData = [];

# 50 actually positive
for i in range(50):
    actualData.append(1);

# then 500 - 50 will be actually negetive
for i in range(450):
    actualData.append(0);

# True Positive
for i in range(45):
    predictedData.append(1);
    
# False Negative
for i in range(5):
    predictedData.append(0);
    
# False Positive
for i in range(55):
    predictedData.append(1)

# True Negative
for i in range(395):
    predictedData.append(0)
```

```python
cfm = confusion_matrix(actualData, predictedData)

# Accuracy (TP + TN)/TOTEL
print("Accuracy Score: ",accuracy_score(actualData,predictedData))

# Precision TP/(TP+FP)
print("Precision Score: ", precision_score(actualData,predictedData))

# Recall
print("Recall Score: ", recall_score(actualData,predictedData))

# F-1 score (2 x pre x recall)/(pre + recall)
print("F-1 Score: ", f1_score(actualData,predictedData))
```
