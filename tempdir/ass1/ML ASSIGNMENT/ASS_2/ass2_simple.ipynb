```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt;
from sklearn import linear_model
import math
```

```python
df = pd.read_csv("temperatures.csv")
df.head(5)
```

```python
col = df.loc[ : , "JAN":"DEC"]
```

```python
col
```

```python
df['AVERAGE'] = col.mean(axis=1)
df.head(5)
```

```python
data = df[['YEAR','AVERAGE']]
data
```

```python
data.shape
```

```python
train_df = data[0:100]
```

```python
train_df.head # for train
```

```python
temp_df = data[["YEAR"]] # for test
test_df = temp_df[101:117]
test_df.head
```

```python
reg = linear_model.LinearRegression()
reg.fit(train_df[['YEAR']],train_df.AVERAGE) # (DataFrame, series)
```

```python
print(data['AVERAGE'][df['YEAR'] == 2003])
print(data['AVERAGE'][df['YEAR'] == 2004])
```

```python
print(reg.predict([[2003]]))
print(reg.predict([[2004]]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
p = reg.predict(test_df)
```

```python
test_df['PREDICTED TEMP'] = p
```
